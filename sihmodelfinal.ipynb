```python
image_path = '/content/nischay aadhar.jpg'
# Extract text from the image
extracted_text = pytesseract.image_to_string(Image.open(image_path))
# Regex to find Aadhaar number (12 digits in groups of 4)
aadhar_number = re.findall(r'\b\d{4}\s\d{4}\s\d{4}\b', extracted_text)
# Check the extracted Aadhaar number
if aadhar_number:
    print(f"Aadhaar Number: {aadhar_number[0]}")
else:
    print("Aadhaar number not found.")
```
